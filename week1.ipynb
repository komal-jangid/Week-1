```python
plt.figure(figsize=(10,6))
sns.histplot(df["generated_power_kw"],bins=30,kde=True)
plt.title("Distribution of Power (kw)")
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["generated_power_kw"],bins=10,kde=False)
plt.title("Distribution of Power (kw)")
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["angle_of_incidence"],bins=30,kde=True)
plt.title("Distribution of Incidence Angle")
plt.xlabel('Incidence Angle)')
plt.ylabel('Frequency')
plt.show()
```
